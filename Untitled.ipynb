```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
```

```python
fig, ax = plt.subplots(figsize=(8,6))

for i in range(len(feature_names)):
    for j in range(len(feature_names)):
        if i != j:
            ax.scatter(X[:,i], X[:,j], alpha=0.5, label=feature_names[j])
            
ax.legend()
ax.set_xlabel('Features')
ax.set_ylabel('Features')
ax.set_title('Scatter plot of Iris data')
plt.show()
```

```python
# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)
```

```python
# perform PCA on iris data
pca = PCA(n_components=3)
pca_transformed = pca.fit_transform(iris.data)

# create 3D plot of PCA-transformed iris data
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_transformed[:, 0], pca_transformed[:, 1], pca_transformed[:, 2], c=iris.target, cmap='viridis')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_title('PCA transformed Iris data')
plt.show()
```
